## Exercice 2 : Interconnexion avec une base de données relationnelle (MySQL)

#### **Installer la bibliothèque nécessaire**



```python
pip install mysql-connector-python pandas
```

#### **code Python pour importer student_performance.csv dans MySQL**

```python
import pandas as pd
import mysql.connector

# Étape 1 : Charger le CSV avec pandas
df = pd.read_csv("student_performance.csv")

# Étape 2 : Connexion à la base de données MySQL
conn = mysql.connector.connect(
    host="localhost",       # ou l'adresse de votre serveur MySQL
    user="root",  
    password="",
    database="etudients"   # nom de votre base de données
)

cursor = conn.cursor()

# Étape 3 : Créer la table "etudiants" si elle n'existe pas
cursor.execute("""
CREATE TABLE IF NOT EXISTS etudiants (
    id INT AUTO_INCREMENT PRIMARY KEY,
    nom VARCHAR(100),
    prenom VARCHAR(100),
    age INT,
    genre VARCHAR(10),
    score_math FLOAT,
    score_lecture FLOAT,
    score_ecriture FLOAT
)
""")

# Étape 4 : Insérer les données du CSV dans la table
for _, row in df.iterrows():
    cursor.execute("""
    INSERT INTO etudiants (nom, prenom, age, genre, score_math, score_lecture, score_ecriture)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
    """, tuple(row))

# Étape 5 : Valider et fermer la connexion
conn.commit()
cursor.close()
conn.close()

print("Importation terminée avec succès !")
```

#### Étape 2 : Vérifier le fichier robots.txt du site

#### Exemple avec le site BBC News.

```python
import requests

# robots.txt
url = "https://www.bbc.com/robots.txt"

response = requests.get(url)

print("Contenu du fichier robots.txt :\n")
print(response.text)
```

#### Étape 3 : Extraire des titres HTML

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.bbc.com/news"

page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")

# récupérer les titres h1
titres = soup.find_all("h1")

print("Titres trouvés :\n")

for t in titres:
    print(t.text)
```

## 2_Consommation d’API avec requests

On peut utiliser l’API gratuite JSONPlaceholder.

#### **Étape 1 : Récupérer les données JSON**

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)

data = response.json()

print(data[:3])   # afficher les 3 premiers résultats
```

#### **Étape 2 : Convertir en DataFrame**

```python

import pandas as pd
import requests

url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)

data = response.json()

df = pd.DataFrame(data)

print(df.head())
```
